# Download GSE277777 data

11/16/2025: Code meant to download GSE277777 data to one directory for analysis

```python
import GEOparse
import os

# Destination directory where everything will be saved
DEST = "."
#
# 1. Download and parse the full GSE record
print("Downloading GSE277777 series...")
gse = GEOparse.get_GEO(
    geo="GSE277777",
    destdir=DEST,
    annotate_gpl=False,     # faster; set True if you need platform tables too
)

# 2. Download all GSM supplementary files
print("Downloading GSM supplementary files...")
for gsm_id, gsm in gse.gsms.items():
    print(f"--> {gsm_id}")
    try:
        gsm.download_supplementary_files(DEST)
    except Exception as e:
        print(f"   Failed for {gsm_id}: {e}")

print("Done.")
```
